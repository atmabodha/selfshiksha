```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df.columns=df.iloc[0]
df=df.iloc[1:]
# df=df.set_index(df.columns[0])
```

```python
df.head()
```

```python
x_list = [float(item) for item in list(df['x'])]
y_list = [float(item) for item in list(df['y'])]

plt.scatter(x_list, y_list,s=0.4)
plt.tick_params(left = False, bottom = False)
plt.show()
```

## Simple Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
regr = LinearRegression()

x_array = np.array(x_list).reshape(-1, 1)
y_array = np.array(y_list).reshape(-1, 1)

regr.fit(x_array,y_array)
print("Slope: ",regr.coef_)
print("Intercept: ",regr.intercept_)
```

```python
y_pred_array = regr.predict(x_array)
y_pred_list = list(y_pred_array.reshape(1,-1)[0])

plt.scatter(x_list, y_list,s=0.4)
plt.plot(x_list, y_pred_list)
plt.tick_params(left = False, bottom = False)
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error

mae = mean_absolute_error(y_true=y_array,y_pred=y_pred_array)
#squared True returns MSE value, False returns RMSE value.
mse = mean_squared_error(y_true=y_array,y_pred=y_pred_array) #default=True
rmse = mean_squared_error(y_true=y_array,y_pred=y_pred_array,squared=False)
  
print("MAE:",mae)
print("MSE:",mse)
print("RMSE:",rmse)
```

## Multiple Linear Regression

```python
x1 = [int(item) for item in list(df['x'])]
y = [float(item) for item in list(df['y'])]
x2 = [item**2 for item in x1]
```

```python
df['x1'] = x1
df['x2'] = x2
df['y'] = y
```

```python
df.drop('x',axis=1,inplace=True)
```

```python
df
```

```python
X = df.drop('y',axis=1)
Y = df['y']
```

```python
regr = LinearRegression()

regr.fit(X,Y)
# print("Slope: ",regr.coef_)
# print("Intercept: ",regr.intercept_)

# y_pred_array = regr.predict(x_array)
```

```python
print("Slope: ",regr.coef_)
print("Intercept: ",regr.intercept_)
```

```python
y_pred_array = regr.predict(X)
y_pred_list = list(y_pred_array.reshape(1,-1)[0])
Y_list = list(Y)
Y2 = [regr.intercept_ + regr.coef_[0]*item + regr.coef_[1]*item**2 for item in x1]

plt.scatter(x1, Y_list,s=0.4)
plt.plot(x1, y_pred_list)
plt.plot(x1, Y2)
plt.tick_params(left = False, bottom = False)
plt.show()
```
