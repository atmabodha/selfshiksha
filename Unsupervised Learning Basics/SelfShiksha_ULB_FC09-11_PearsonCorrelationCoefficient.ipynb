```python
import numpy as np
import random
import plotly.express as px
```

# Positive Correlation

```python
# Create an array of 100 numbers from 0 to 10
x1=np.linspace(0,10,100)

# Create an array of 100 random numbers to be added as noise
noise_magnitude = 10.0
randarray = (np.random.rand(100)-0.5)*noise_magnitude

# Create an array of 100 numbers using x1 to act as the second variable
x2=5.0*x1+3+randarray
```

```python
fig = px.line(x=x1,y=x2)
fig.show()
```

```python
np.corrcoef(x1,x2)
# This outputs a 2D array which is called the Covariance Matrix, whose values range between -1.0 to 1.0
# A value of 1.0 indicates perfect positive correlation and a value of -1.0 indicates perfect negative correlation.
# element [0,0] of this 2D array is the correlation of x1 with itself, which will obviously be 1.0
# element [1,1] of this 2D array is the correlation of x2 with itself, which will obviously be 1.0
# elements [0,1] and [1,0] will be the same and indicate the correlation between x1 and x2
```

```python
# For the data above, the correlation between x1 and x2 is ~ 0.979, which indicates a very high positive correlation.
np.corrcoef(x1,x2)[0][1]
```

```python
np.corrcoef(x1,x2)
```
