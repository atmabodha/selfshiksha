```python
import pandas as pd
import numpy as np

from skimage.io import imread,imshow

from sklearn.cluster import MiniBatchKMeans

import plotly.express as px
import plotly.graph_objs as go
import matplotlib.pyplot as plt
```

```python
dog = imread("SelfShiksha_ULB_FC65_Dog.png")
```

```python
type(dog)
```

```python
dog.shape
```

```python
ax = plt.axes(xticks=[], yticks=[])
ax.imshow(dog)
```

```python
# The last dimension has redundant values which can be dropped!
ax = plt.axes(xticks=[], yticks=[])
ax.imshow(dog[:,:,3])
```

```python
dog = dog[:,:,0:3]
```

```python
doggy = dog / 255.0 # use 0...1 scale
doggy = doggy.reshape(dog.shape[0] * dog.shape[1], dog.shape[2])
doggy.shape
```

```python
kmeans = MiniBatchKMeans(16)
kmeans.fit(doggy)
new_doggy = kmeans.cluster_centers_[kmeans.predict(doggy)]
new_dog = new_doggy.reshape(dog.shape)

ax = plt.axes(xticks=[], yticks=[])
ax.imshow(new_dog)
```

```python
kmeans = MiniBatchKMeans(64)
kmeans.fit(doggy)
new_doggy = kmeans.cluster_centers_[kmeans.predict(doggy)]
new_dog = new_doggy.reshape(dog.shape)

ax = plt.axes(xticks=[], yticks=[])
ax.imshow(new_dog)
```
